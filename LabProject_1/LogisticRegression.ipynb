```python
# -*- coding: utf-8 -*-
"""
Created on Sun Sep 17 09:58:48 2017

@author: Preeti
"""
import pandas as pd
import seaborn as sb
import numpy as np

dataset = pd.read_csv('C:/DataScience/Stats3/OnlineNewsPopularity.csv') 

dataset = dataset.drop(['shares','weekday_is_monday','weekday_is_tuesday','weekday_is_wednesday','weekday_is_thursday','weekday_is_friday','weekday_is_saturday','weekday_is_sunday','is_weekend'],1)
dataset = dataset.drop(['data_channel_is_lifestyle','data_channel_is_entertainment','data_channel_is_bus','data_channel_is_socmed','data_channel_is_tech','data_channel_is_world'],1)
dataset = dataset.drop(['ID'],1)


x = dataset.iloc[:, :-1].values
# iloc[rows,columns] function : means all. all rows and all columns except last

y = dataset['popularity']

#### another co-relation plots    
corr=dataset.corr()
sb.heatmap(corr)

# get positions of columns to be converted from string to int
idx_day = dataset.columns.get_loc('day')
idx_channel = dataset.columns.get_loc('channel')


# Encoding the Independent Variable
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_X = LabelEncoder()
x[:,idx_day] = labelencoder_X.fit_transform(x[:,idx_day])
x[:,idx_channel] = labelencoder_X.fit_transform(x[:,idx_channel])
onehotencoder = OneHotEncoder(categorical_features = [idx_day,idx_channel])
x = onehotencoder.fit_transform(x).toarray()
# Encoding the Dependent Variable
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)

from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=0.25,random_state=0)

## logistic regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train,y_train)

y_pred = classifier.predict(x_test)


## confusion matrix
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
from sklearn.metrics import log_loss

cm = confusion_matrix(y_test, y_pred)

print('Logistic regression results:')
print('=============================')
print('Classification summary')
print(classification_report(y_test, y_pred))
print('------------------------------------------------')
print('accuracy: ',cross_val_score(classifier, x_test, y_test, cv=10).mean())
print('------------------------------------------------')
print('Confusion matrix: ')
print(cm)
print('------------------------------------------------')
print('Co-efficients')
print(np.exp(classifier.coef_.T))
print('*********************************************************')
```
